```python
# read in the dataset
def get_movies(aDir):
    import os, json
    file_contents = os.listdir(aDir)

    movie_list = []

    for filename in file_contents:
        filepath = os.path.join(aDir, filename)

        if not filename.endswith(".json"):
            continue
        
        with open(filepath, 'r') as movie_file:
            movie_data = json.load(movie_file)
        if hasattr(movie_data, "keys"): # type(movie_data) == dict:
            movie_list.append(movie_data)
    return movie_list
```

```python
# replacing NaN in a column to a list
def NaNtolist(elt):
    if type(elt)==list:
        return elt
    else:
        return ['z']

critics['genre'] = [NaNtolist(elt) for elt in critics['genre']]
```

```python
# replacing NaN in a column to a list
def tolist(elt):
    if type(elt)==list:
        return elt
    else:
        return [0,0,0,0]

critics['num_critic_reviews'] = [tolist(elt) for elt in critics['num_critic_reviews']]
critics['num_user_reviews'] = [tolist(elt) for elt in critics['num_user_reviews']]
```

```python
# convert reviews list to columns in the dataframe - num_critic_reviews , num_user_reviews
critics.loc[:, 'pos_critic_reviews'] = [rev[0] for rev in critics.num_critic_reviews]
critics.loc[:, 'neu_critic_reviews'] = [rev[1] for rev in critics.num_critic_reviews]
critics.loc[:, 'neg_critic_reviews'] = [rev[2] for rev in critics.num_critic_reviews]
critics.loc[:, 'tot_critic_reviews'] = [rev[3] for rev in critics.num_critic_reviews]

critics.loc[:, 'pos_user_reviews'] = [rev[0] for rev in critics.num_user_reviews]
critics.loc[:, 'neu_user_reviews'] = [rev[1] for rev in critics.num_user_reviews]
critics.loc[:, 'neg_user_reviews'] = [rev[2] for rev in critics.num_user_reviews]
critics.loc[:, 'tot_user_reviews'] = [rev[3] for rev in critics.num_user_reviews]
```

```python
import statsmodels.api as sm

X = sm.add_constant(df_model[['f_recency',
  'f_title_length',
 'f_more_movie_directed',
  'f_R_rated',
  'f_PG13_rated',
  'f_not_rated',
  'f_PG_rated',
 'f_log_open_theater',
  'f_log_open_wkend',
  'f_square_log_wkend',
  'f_per_theater_take',
   'f_action' ,
    'f_neg_user_percent',
    'f_pos_user_percent',
    'f_neg_critic_percent',
    'f_pos_critic_percent']])
Y = df_model['f_log_domestic']
```

```python
from sklearn.cross_validation import train_test_split

# splits x -> x_train, x_test
#        y -> y_train, y_test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size=0.4)

print len(X_train), len(X_test)
```

```python
model = sm.OLS(Y_train, X_train).fit()
model.summary()
```

```python
from sklearn.metrics import mean_squared_error

# Root mean squared error is standard deviation of
# the differences between predicted and actual values
def RMSE(model, X_, Y_):
    Y_pred = model.predict(X_)
    Y_true = Y_
    MSE = mean_squared_error(Y_pred, Y_true)
    return np.sqrt(MSE)

train_RMSE = RMSE(model, X_train, Y_train)
test_RMSE  = RMSE(model, X_test, Y_test)

print 'Training RMSE is $%.1f dollar' % (train_RMSE /1e0)
print 'Test RMSE     is $%.1f dollar' % (test_RMSE / 1e0)
```
